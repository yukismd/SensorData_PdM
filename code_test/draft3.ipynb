```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data_modified/modeling_engine.csv')
df.shape
```

```python
df
```

```python
df.describe()
```

```python
df['time_to_deterioration'].quantile([i/10 for i in range(11)])
```

```python
df['time_to_deterioration'].plot(kind='hist', grid=True, bins=40)

plt.xlabel('Time to deterioration')  # x軸ラベル
plt.title('Distribution of time to deterioration')  # グラフのタイトル
plt.show()
```

```python
pd.DataFrame(df['time_to_deterioration'].apply(np.log)).plot(kind='hist', grid=True, bins=40)

plt.xlabel('Time to deterioration (log)')  # x軸ラベル
plt.title('Distribution of time to deterioration (log)')  # グラフのタイトル
plt.show()
```

```python
# sampling
```

```python
df.shape
```

```python
df_sample = df.sample(frac=0.1, random_state=123)
df_sample = df_sample.sort_values(['unit_ID', 'cycles'])
df_sample.shape
```

```python
df_sample
```

```python
df_sample['time_to_deterioration'].plot(kind='hist', grid=True, bins=40)

plt.xlabel('Time to deterioration')  # x軸ラベル
plt.title('Distribution of time to deterioration')  # グラフのタイトル
plt.show()
```

```python
pd.DataFrame(df_sample['time_to_deterioration'].apply(np.log)).plot(kind='hist', grid=True, bins=40)

plt.xlabel('Time to deterioration (log)')  # x軸ラベル
plt.title('Distribution of time to deterioration (log)')  # グラフのタイトル
plt.show()
```

```python
df_sample.to_csv('data_modified/modeling_engine_partial.csv', index=False)
```

```python
# train, test 
```

```python
df_sample['unit_ID'].unique()
```

```python
np.random.seed(seed=123)
test_unit = np.random.choice(df_sample['unit_ID'].unique(), size=26, replace=True)
test_unit.sort()
test_unit
```

```python
df_train = df_sample[~df_sample['unit_ID'].isin(test_unit)]
df_test = df_sample[df_sample['unit_ID'].isin(test_unit)]
df_train.shape, df_test.shape
```

```python
df_train['time_to_deterioration'].plot(kind='hist', grid=True, bins=40)

plt.xlabel('Time to deterioration')  # x軸ラベル
plt.title('Distribution of time to deterioration (Train)')  # グラフのタイトル
plt.show()
```

```python
df_test['time_to_deterioration'].plot(kind='hist', grid=True, bins=40)

plt.xlabel('Time to deterioration')  # x軸ラベル
plt.title('Distribution of time to deterioration (Test)')  # グラフのタイトル
plt.show()
```

```python
column_names = ['unit_ID','cycles','setting_1','setting_2','setting_3','T2','T24','T30','T50','P2','P15','P30','Nf',
           'Nc','epr','Ps30','phi','NRf','NRc','BPR','farB','htBleed','Nf_dmd','PCNfR_dmd','W31','W32' ]
sensor_column_names =  ['setting_1','setting_2','setting_3','T2','T24','T30','T50','P2','P15','P30','Nf',
           'Nc','epr','Ps30','phi','NRf','NRc','BPR','farB','htBleed','Nf_dmd','PCNfR_dmd','W31','W32' ]
len(sensor_column_names)
```

```python
tmp = df[df['unit_ID']==1]
tmp
```

```python
df_list_y = []
df_list_X = []

cycle_range = range(101, 201)    # 予測を実施するcycle範囲
past_cycle_search = 25   # 現cycleから見て、どれくらい過去cycleまで取得するか

for cyc in cycle_range:
    #print(cyc)
    #print(tmp[tmp['cycles']==cyc]['time_to_deterioration'])
    if tmp[tmp['cycles']==cyc]['time_to_deterioration'].values[0] <= 0:    # 現cycleのtime_to_deteriorationが0となるとそのunitは終了
        break
    df_list_y.append(tmp[tmp['cycles']==cyc])
    
    # -25までのデータ
    past_cycle = range(cyc-past_cycle_search, cyc)
    #print(past_cycle)
    X_for_cyc = tmp[tmp['cycles'].isin(list(past_cycle))]
    assert X_for_cyc.shape[0] == past_cycle_search
    df_list_X.append(X_for_cyc)

len(df_list_y), len(df_list_X)
```

```python
df_list_y = []
df_list_X = []

cycle_range = range(101, 201)    # 予測を実施するcycle範囲
past_cycle_search = 25   # 現cycleから見て、どれくらい過去cycleまで取得するか

test_count = 0

for unit in df['unit_ID'].unique():     # unit_IDのループ
    d_unit = df[df['unit_ID']==unit]
    #print(unit, d_unit.shape)

    for cyc in cycle_range:     # 各unit_ID内、対象cycleのループ
        #print(cyc)
        if d_unit[d_unit['cycles']==cyc]['time_to_deterioration'].values[0] <= 0:    # 現cycleのtime_to_deteriorationが0となるとそのunitは終了
            break
        df_list_y.append(d_unit[d_unit['cycles']==cyc])   # 現時点のcycle

        past_cycle = range(cyc-past_cycle_search, cyc)
        #print(past_cycle)
        X_for_cyc = d_unit[d_unit['cycles'].isin(list(past_cycle))]   # 現時点から見た、過去分のcycle
        assert X_for_cyc.shape[0] == past_cycle_search
        df_list_X.append(X_for_cyc)
        test_count += 1

test_count, len(df_list_y), len(df_list_X)
```

```python
df_modeling = pd.DataFrame()
for dfy, dfX in zip(df_list_y, df_list_X):
    y = dfy[['unit_ID','cycles','time_to_deterioration']]
    X = dfX[sensor_column_names]
    X_mean = pd.DataFrame(X.mean()).T
    X_mean.columns = [nm+'-mean' for nm in sensor_column_names]
    X_sd = pd.DataFrame(X.std()).T
    X_sd.columns = [nm+'-sd' for nm in sensor_column_names]
    d = pd.concat([y.reset_index(drop=True), X_mean.reset_index(drop=True), X_sd.reset_index(drop=True)], axis=1)
    #display(d)
    assert d.shape == (1, 51)
    df_modeling = pd.concat([df_modeling, d])
    #display(df_modeling)
df_modeling.shape
```
